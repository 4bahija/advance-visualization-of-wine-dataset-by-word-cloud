```python
import pandas as pd
import numpy as np
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator
from os import path
from PIL import Image
import matplotlib.pyplot as plt
```

```python
wine=pd.read_csv('winemag-data_first150k.csv')
wine.head(10)
```

```python
print("There are {} observations and {} features in this dataset\n".format(wine.shape[0],wine.shape[1]))
print("There are {} varities of wine in this dataset such as {}.....\n".format(len(wine.variety.unique()),",".join(wine.variety.unique()[0:5])))
print("There are {} countries producing wine in this dataset such as {}.....\n".format(len(wine.country.unique()),",".join(wine.country.unique()[0:5])))
```

```python
wine[["country","description","points"]].head()
```

```python
country=wine.groupby("country")
country.describe().head()
```

```python
country.mean().sort_values(by="points",ascending=False).head()
```

```python
plt.figure(figsize=(10,10))
country.size().sort_values(ascending=False).plot.bar()
plt.xticks(rotation=50)
plt.xlabel("country of origin")
plt.ylabel("number of wines")
plt.show()
```

```python
text=wine.description[0]
wordcloud=WordCloud(max_font_size=100).generate(text)
plt.imshow(wordcloud,interpolation='bilinear')
plt.axis("off")
plt.show()
```
